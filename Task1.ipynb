```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('world-population-by-country-2020.csv')
df
```

```python
df.columns
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# Convert column to numeric
df['Migrants (net)'] = pd.to_numeric(df['Migrants (net)'], errors='coerce')

# Fill missing values
df['Migrants (net)'] = df['Migrants (net)'].fillna(df['Migrants (net)'].median())
```

```python
df['Migrants (net)'].isnull().sum()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Med. Age'], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```
